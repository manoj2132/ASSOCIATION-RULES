#What is lift and why is it important in Association rules?
ans. Lift is a metric that measures the strength of an association between two items in association rule mining. It compares the observed co-occurrence of two items with their expected co-occurrence if they were independent.

Importance:
Lift > 1: The items are more likely to occur together than by chance (positive association).
Lift = 1: The items are independent of each other (no association).
Lift < 1: The items are less likely to occur together than by chance (negative association).
Lift helps identify the strength and direction of relationships, making it crucial for finding meaningful patterns.

#What is support and Confidence. How do you calculate them?
Support and Confidence are key metrics in association rule mining:
Support is the frequency an itemset appears, and Confidence measures the reliability of a rule. They are calculated as:
Support = (Frequency of itemset) / (Total transactions)
Confidence = (Frequency of itemset) / (Frequency of antecedent)
#What are some limitations or challenges of Association rules mining?
Some limitations or challenges of association rule mining include:

Scalability: Handling large datasets with many items can lead to computational challenges and slow performance.
Too many rules: It can generate an overwhelming number of rules, making it difficult to identify useful patterns.
Low interpretability: Some rules might be statistically strong but lack real-world significance.
Rare item problem: Rare but important associations may not be discovered due to low support thresholds.
